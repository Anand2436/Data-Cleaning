Doing Titanic data set Cleaning

```python
import numpy as np
import pandas as pd
```

```python
tit = pd.read_csv("titanic_train.csv")

#print(tit.head())
print(tit.shape)

tit
```

```python
tit1 = tit.copy()

def gender(str):
    if str == "male":
        return 1
    else:
        return 2
    
tit1["Gender"] = tit1.Sex.apply(gender)
tit1.shape

del tit1["Sex"]
tit1
```

```python
#Deleting unrelevant data

del tit1["PassengerId"]
del tit1["Name"]
del tit1["Ticket"]

print(tit1.shape)
```

# Now our task is to replace nan in age column
==> They must be replaced acc to following t and cs
 1.The nans corresponding to survived person should be replaced by means of survived people ages
 2.The nans corresponding to not-survived person should be replaced by means of not-survived people ages


```python
print(tit1.isnull().sum())
tit1.describe()
```

==> Finding the number of people who have survived , given that they have embarked or boarded from a particular port


```python
# Port S : Checking if ports can be used comprehensively to predict survival rate 

Sur_S = tit1[tit1.Embarked =="S"][tit1.Survived == 1].shape[0]
NSur_S = tit1[tit1.Embarked =="S"][tit1.Survived == 0].shape[0]
print("S Port Results (1 0):",Sur_S,NSur_S)

# Port C
Sur_C = tit1[tit1.Embarked =="C"][tit1.Survived == 1].shape[0]
NSur_C = tit1[tit1.Embarked =="C"][tit1.Survived == 0].shape[0]
print("C Port Results (1 0):",Sur_C,NSur_C)

# Port Q
Sur_Q = tit1[tit1.Embarked =="Q"][tit1.Survived == 1].shape[0]
NSur_Q = tit1[tit1.Embarked =="Q"][tit1.Survived == 0].shape[0]
print("Q Port Results (1 0):",Sur_Q,NSur_Q)
```

```python
# Plotting some data 
import matplotlib.pyplot as plt

surCount = (tit1.Survived == 1).sum()
NsurCount = (tit1.Survived == 0).sum()
# Same as NsurCount = (tit1.shape[0]-surCount)

p = [surCount, NsurCount]

label = ["Survived","Not Survived"]

exp = [0.09,0]

color = ["black","red"]

plt.pie(p,labels = label,colors = color,explode = exp,startangle = 90)
plt.axis("equal")
plt.show()
```

```python
# Finding mean survived age

meanSur = tit1[tit1.Survived == 1].Age.mean()
meanSur
```

```python
# Creating new age column to replace nan having survived == 1, replaced by mean survived age

tit1["age"] = np.where(tit1.Age.isnull() & tit1.Survived == 1, meanSur , tit1["Age"])
# syntax : np.where(condition,val1,val2)
# if condition == True, put val1 else val2
tit1
```

```python
# calculating mean of ages of Not Survived

meanNsur = tit1[tit1.Survived==0].Age.mean()
meanNsur
```

```python
# Now that only Nans left in our new age column are of not survived
# Therefore we neednot use where

tit1.age.fillna(meanNsur,inplace = True)
tit1
```

```python
# Remove the old age column and rename new age column to the one that existed 
# 1. del "Age" and change "age" == "Age"

del tit1["Age"]
```

```python
tit1.rename(columns ={"age":"Age","Gender":"Sex"},inplace = True)
tit1
```

```python
# Now cabin has some nans

tit1.dropna(inplace=True)
tit1
```
